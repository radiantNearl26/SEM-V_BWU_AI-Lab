```python
import pandas as pd

df = pd.DataFrame({
    'Models': ['GPT-5.6', 'Opus-4.8', 'Kimi-v2.6', 'Qwen-v4'],
    'Accuracy': [92.5,93.7,98.1,95.4]
})

display(df)

print("\nModels with high accuracy: ")
high_perf = df[df['Accuracy'] > 95]
display(high_perf)
```

```python
from matplotlib import colors
import matplotlib.pyplot as plt

epochs = [1,2,3,4,5,6,7,8,9]
train_loss = [0.9,0.4,0.7,0.9,0.9,0.4,0.7,0.9,0.1]
val_loss = [0.95,0.91,0.94,0.98,0.96,0.95,0.91,0.98,0.99]

plt.figure(figsize=(8,6))
plt.plot(epochs, train_loss, label='Training Loss', color='royalblue', marker='o')
plt.plot(epochs, val_loss, label='Validation Loss', color='tomato', marker='*')

plt.xlabel('Epochs')
plt.ylabel('Loss')

plt.title('Training and Validation Loss')
plt.legend()

plt.show()
```

# Extract elements using boolean masking

```python
matrix = np.random.randint(1,16,(4,4))
print(matrix)
```

```python
extracted = matrix[(matrix > 10) & (matrix % 2 == 0)]
print(extracted)
```

# Side by Side subplots

```python
epochs = np.linspace(1,5,5)
accuracy = np.random.randint(90,99,(5))

print(epochs)
print(accuracy)
```

```python
inf_speed = pd.DataFrame({
    'machine': ['A','B','C'],
    'inf_speed': [19,54,82]
})

display(inf_speed)
```

```python
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.plot(epochs, accuracy, label='Accuracy trend over epochs', color='royalblue')

plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.title('Training Accuracy')

plt.subplot(1,2,2)
plt.bar(data=inf_speed, x='machine', height='inf_speed', color='royalblue')

plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.title('Validation Accuracy')

plt.show()
```
